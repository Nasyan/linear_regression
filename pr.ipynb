Predict the price based on the number of sales. Polynomial regression.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Generate data:

```python
np.random.seed(2)
X = np.random.normal(3.0,1.0,1000) # (mean, dispersion, quantity)
y = np.random.normal(10.0,1.0,1000) / X # (mean, dispersion, quantity)
```

Data visualization:

```python
plt.scatter(X, y)
plt.xlabel('Sales')
plt.ylabel('Price')
plt.show()
```

Setting the polynomial degree:

```python
degree = 4
```

Add column with ones:

```python
X_with_ones = np.ones((len(X), degree+1))
for i in range(1, degree+1):
    X_with_ones[:, i] = X**i

w = np.linalg.inv(X_with_ones.T.dot(X_with_ones)).dot(X_with_ones.T).dot(y)
```

Generate line:

```python
X_plot = np.linspace(0, 7, num=100)
y_plot = np.zeros(len(X_plot))
for i in range(degree+1):
    y_plot += w[i] * X_plot**i
```

Data vizualization:

```python
plt.scatter(X, y)
plt.xlabel('Sales')
plt.ylabel('Price')
plt.plot(X_plot, y_plot, color='red')
plt.show()
```

Prediction:

```python
X_new = np.array([[4]]) # new value
X_new_with_ones = np.ones((len(X_new), degree+1))
for i in range(1, degree+1):
    X_new_with_ones[:, i] = X_new**i

y_pred = X_new_with_ones.dot(w)
print('Predicted price:', y_pred)
```

MSE:

```python
y_pred = X_with_ones.dot(w)
mse = np.mean((y - y_pred) ** 2)
print(f'MSE: {mse:.4f}')
```
