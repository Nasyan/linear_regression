Predict students' grades based on the time spent on preparation. Analytical solution.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read data:

```python
data = pd.read_csv('student_scores.csv')
X = data['Hours'].values.reshape(-1, 1)
y = data['Scores'].values
```

Data visualization:

```python
plt.scatter(X, y)
plt.xlabel('Study Time')
plt.ylabel('Scores')
plt.show()
```

Linear regression solution:

```python
X_with_ones = np.concatenate((np.ones((len(X), 1)), X), axis=1) # add column with ones (for free term)
w = np.linalg.inv(np.dot(X_with_ones.T, X_with_ones)).dot(X_with_ones.T).dot(y) # calculation
```

Result visualization:

```python
plt.scatter(X, y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.plot(X, np.dot(X_with_ones, w), color='red')
plt.show()
```

Prediction:

```python
X_new = np.array([[4]]) # new value
X_new_with_ones = np.concatenate((np.ones((len(X_new), 1)), X_new), axis=1) # add column with ones (for free term)
y_pred = np.dot(X_new_with_ones, w) # calculating predicted score
print('Predicted score:', y_pred)
```

MSE:

```python
mse = np.mean((y - np.dot(X_with_ones, w)) ** 2) # calculation
print(f'MSE: {mse:.4f}')
```
